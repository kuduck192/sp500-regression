```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
import wikipedia as wp
import yfinance as yf 
from ta.momentum import rsi 
from ta.trend import macd
```

```python
valid_symbols = pd.read_csv('../input/clear_domain.csv')['Symbol'].unique().tolist()
len(valid_symbols)
```

```python
stock_prices = [
    yf.Ticker(symbol).history(period="15y", auto_adjust=False) for symbol in valid_symbols
]

stock_prices
```

```python
sp500_df = pd.concat(stock_prices, keys=valid_symbols)
sp500_df
```
